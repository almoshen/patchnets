```python
import json

with open('examples/splits/sv1_generalization50_test.json') as f:
    split = json.load(f)

import os
import shutil
for dataset in split:
  for class_name in split[dataset]:
    for instance_name in split[dataset][class_name]:
      full_file_name = os.path.join(f'experimentresults/patchnets_generalization_train_on_50_per_category_test_shapenet/output/parts/parts_mixture/latest/ShapeNetV2/{class_name}', instance_name)
      # full_file_name = os.path.join(f'/mnt/g/Shared drives/Dataset_ShapeNetCore/data/ShapeNetCore.v2/{class_name}', instance_name[:-4] + '/models/model_normalized.obj')
      if os.path.isfile(full_file_name):
        recondir = f'experimentresults/patchnets_generalization_train_on_50_per_category_test_shapenet/output/reconstruction/{class_name}'
        if not os.path.isdir(recondir):
          os.mkdir(recondir)
        target = os.path.join(recondir,instance_name)
        if os.path.isfile(target):
          print(f'{instance_name} already exists')
        else:
          shutil.copy(full_file_name, target)
```
